```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
print(x)
# print(x.astype(int))
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('square function poting')
plt.show()
```

```python

# multiplot
# plt.subplot(rows, cols, plot_number(position))
plt.subplot(1, 1, 1)
plt.plot(x, x, 'b')

plt.subplot(1, 1, 1)
plt.plot(x, x**2, 'r')

plt.subplot(1, 1, 1)
plt.plot(x, x**3, 'g')

plt.subplot(1, 1, 1)
plt.plot(x, x**4, 'c')

plt.subplot(1, 1, 1)
plt.plot(x, x**5, 'm')

plt.show()
```

```python
fig, ax = plt.subplots()

fig.set_size_inches(10, 5)
fig.suptitle("Multiple Functions on One Plot")

ax.plot(x, x,     'b', label='x')
ax.plot(x, x**2,  'r', label='x^2')
ax.plot(x, x**3,  'g', label='x^3')
ax.plot(x, x**4,  'c', label='x^4')
ax.plot(x, x**5,  'm', label='x^5')

ax.set_xlabel("X values")
ax.set_ylabel("Y values")
ax.legend()

# fig.tight_layout()

plt.show()
```

```python
fig, ax = plt.subplots()

fig.set_size_inches(10, 5)
fig.suptitle("Multiple Functions on One Plot")

ax.plot(x, x,     'b', label='x')
ax.plot(x, x**2,  'r', label='x^2')
ax.plot(x, x**3,  'g', label='x^3')
ax.plot(x, x**4,  'c', label='x^4')
ax.plot(x, x**5,  'm', label='x^5')

ax.set_xlabel("X values")
ax.set_ylabel("Y values")
ax.legend()

fig.tight_layout()

plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
# now we have an empty canvas and we can plot on it

axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')

fig.tight_layout()

plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
fig.set_size_inches(10, 5)

axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')

axes.plot(x, x)
# fig.tight_layout()

plt.show()
```

```python

# we can add more than one axes to the canvas
fig = plt.figure()
# the first axes
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes1.plot(x,y)
axes1.set_title('Larger Plot')

# the second axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
axes2.plot(y,x)
axes2.set_title('Smaller Plot')

plt.show()
```

```python
# subplots() method returns a tuple containing a figure and axes object(s)
fig,axes = plt.subplots() # this is a tuple unpacking

plt.show()
```

```python
# we can specify the number of rows and columns

fig,axes = plt.subplots(nrows=3, ncols=3)

fig.tight_layout() # here, this function is visible

plt.show()
```

```python
# empty canvas
fig,axes = plt.subplots(nrows=1, ncols=2)
fig.set_size_inches(10, 5)

plt.show()
```

```python

# subplots() method takes two arguments, nrows and ncols (number of rows and number of columns)
# it returns a tuple containing a figure and axes object(s)
fig,axes = plt.subplots(nrows=1, ncols=2) # this is a tuple unpacking

# we can iterate through the axes, because it is a numpy array containing two axes
axes[0].plot(x,y)

axes[1].plot(y,x)

plt.show()
```

```python
type(axes)
# print(type(axes))
```

```python
# iterating through the axes
fig,axes = plt.subplots(nrows=1, ncols=2) # this is a tuple unpacking

for current_ax in axes:
    current_ax.plot(x,y)
plt.show()
```

```python

fig = plt.figure(figsize=(3,2,),dpi=300) # figsize is a tuple that specifies the width and height of the figure in inches
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)

plt.show()
```

```python
fig = plt.figure(figsize=(8,2,),dpi=100) # figsize is a tuple that specifies the width and height of the figure in inches
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)

fig.set_size_inches(15,10)

plt.show()
```

```python

# we cna also do the same thing with subplots
fig,axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2,),dpi=100) # figsize is a tuple that specifies the width and height of the figure in inches   

axes[0].plot(x,y)
axes[1].plot(y,x)

# if there is some overlapping between the two plots
plt.tight_layout()

plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color="blue", alpha=0.5)
ax.plot(x, x+2, color="#8B008B")
ax.plot(x, x+3, color="#FF8C00") 

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# possible linestype options -, , -., :, steps
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green")

plt.show()
```
